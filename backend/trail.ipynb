```python
import pandas as pd

# Function to ask a question and get SQL results
def ask_question(question: str):
    sql_query, df, _ = vn.ask(question, print_results=False)

    if df is not None:
        print(f"\nGenerated SQL:\n{sql_query}\n")
        print("Results:")
        print(df.head())
    else:
        print("No results found.")
        df = pd.DataFrame()  # Create an empty DataFrame

    return sql_query, df

# Generate a summary from the results
def generate_summary(question: str, df: pd.DataFrame):
    if df is not None and not df.empty:
        return vn.generate_summary(question, df)
    return "No data available for summary."

# Generate follow-up questions
def generate_followup_questions(question: str, sql_query: str, df: pd.DataFrame):
    if df is not None and not df.empty:
        return vn.generate_followup_questions(question, sql_query, df, n_questions=5)
    return ["No relevant follow-up questions."]

# Determine if a chart should be generated
def should_generate_chart(df: pd.DataFrame):
    return vn.should_generate_chart(df)
```

```python
if __name__ == "__main__":
    user_question = "What was the trend in revenue over the past years?"
    sql_query, result_df = ask_question(user_question)

    # Generate summary
    summary = generate_summary(user_question, result_df)
    print("\nSummary:")
    print(summary)

    # Generate follow-up questions
    followup_qs = generate_followup_questions(user_question, sql_query, result_df)
    print("\nFollow-up Questions:")
    for q in followup_qs:
        print(f"- {q}")

    # Check if a chart should be generated
    if should_generate_chart(result_df):
        print("\nA chart can be generated based on the data.")
    else:
        print("\nA chart is not recommended for this data.")
```
